```python
messages= pd.read_csv("/content/SMSSpamCollection.txt", sep='\t', names=['label','message'])
```

```python
messages.shape
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
stop_words = set(stopwords.words('english'))

def preprocess(text):
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    words = word_tokenize(text)
    words = [word.lower() for word in words]
    words = [word for word in words if word not in stop_words]
    words = [lem.lemmatize(word) for word in words]
    words = [port.stem(word) for word in words]
    return ' '.join(words)
```

```python
sentences=[]
for i in range(len(messages)):
  messages['message'][i]=messages['message'][i].lower()
  pos_tag=nltk.pos_tag(word_tokenize(messages['message'][i]))
  sentences.append(sent_tokenize(messages['message'][i]))

preprocessed_sentences=[]
for sentence_list in sentences:
  preprocessed_sentence_list = []
  for sentence in sentence_list:
    preprocessed_sentence_list.append(preprocess(sentence))
  preprocessed_sentences.append(preprocessed_sentence_list)

preprocessed_sentences
```
